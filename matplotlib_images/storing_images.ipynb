7. We can also flip an image 90 degrees left or right by modifying the pixels of the image. Create a new image that is a 90 degree rotation counterclockwise of the original.

```python

image_cropped2 = image[:,50:175]
width, height = len(image_cropped2[0]), len(image_cropped2)

image_rotated1 = [[image_cropped2[height-1 - x][y] for x in range(height)] for y in range(width)]

# same as above but without list comprehension
image_rotated2 = [[0 for pixel in range(height)] for row in range(width)]

for y in range(width):
    for x in range(height):
        image_rotated2[width - 1 - y][x] = image_cropped2[x][y]


plt.imshow(image_rotated1)
plt.show()
plt.imshow(image_rotated2)
```

8. This is just for fun, but we can look at the red, green, and blue pixels of your image individually using array slicing. You may need to `pip install numpy`.

```python
import numpy as np 

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,8))
 
for c, ax in zip(range(3), axs):
    rgb_img = np.zeros(image.shape, dtype="uint8")
    rgb_img[:,:,c] = image[:,:,c]
    ax.imshow(rgb_img)
    ax.set_axis_off()
```
